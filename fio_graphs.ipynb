```python
import json
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
with open('packet_config_1.json', 'r') as f:
    packet_config = json.load(f)
sizes = [int(s[:-1]) for s in packet_config]
# packet_config
```

```python
# common parameters

common_params_names = ['transmission_time', 'latency_time', 'read_processing_time', 'write_processing_time']

plt.figure(figsize=(6,4))
for param in common_params_names:
    mean = [v[param]['mean'] for v in packet_config.values()]
    plt.plot(sizes, mean, label=param + ' mean')
plt.legend()
plt.grid()
plt.xlabel('block size (KB)')
plt.ylabel('seconds')
plt.title('Fio results: common parameters')
plt.legend(loc='upper left')
plt.show()
```

```python
# read/write parameters

test_names = ['seq_read', 'seq_write', 'rand_read', 'rand_write', 'seq_read_write', 'rand_read_write']
rw_params_names = ['rate_time', 'seek_time', 'overheads_time']

for test in test_names:
    plt.figure(figsize=(6,4))
    for param in rw_params_names:
        mean = [v[test][param]['mean'] for v in packet_config.values()]
        plt.plot(sizes, mean, label=param + ' mean')
    plt.legend()
    plt.grid()
    plt.xlabel('block size (KB)')
    plt.ylabel('seconds')
    plt.title('Fio results: ' + test)
    plt.legend(loc='upper left')
    plt.show()
```
