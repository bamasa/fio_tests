# Гистограмма параметров

* 1 диск
* packet_size = 4K
* runtime = 10 sec
* block_sizq = 4K
* mode = sequential readwrite

```python
import json
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
```

```python
with open('hists/hist_1000.json', 'r') as f:
    res = json.load(f)
```

```python
rw_modes = ['read', 'write']
params = ['latency_time', 'processing_time', 'transmission_time']
```

```python
n_bins = len(res['read']['latency_time']['mean'])
```

```python
kwargs = dict(histtype='stepfilled', alpha=0.3, bins=n_bins)

for param in params:
    plt.figure(figsize=(10,6))
    plt.grid()
    plt.xlabel('sec')
    plt.ylabel('n')
    for rw_mode in rw_modes:
        plt.hist(res[rw_mode][param]['mean'], label=rw_mode, **kwargs)       
    plt.title('Fio results hist: ' + param)
    plt.legend(loc='upper left')
    plt.show()
```
